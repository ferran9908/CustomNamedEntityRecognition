```python
# Test Model
train = ["The patient has a high fever","The patient has acne","Ferran has been infected with h1b1 virus"]
nlp2 = spacy.load('modelTrained')

for text in train:
    doc = nlp2(text)
    print("Entities", [(ent.text, ent.label_) for ent in doc.ents])
    print("Tokens", [(t.text, t.ent_type_, t.ent_iob) for t in doc])
```

```python
medList = []
with open('medList.txt','r') as fileObj:
    for line in fileObj:
        line = line[:-1]
        medList.append(line)
```

```python
medList
```

```python
with open('medicines.txt','w') as fileObj:
    for med in medList:
        fileObj.write('The medicine prescribed for diagnosis is '+ med +'\n')
        fileObj.write('Take 2 doses of ' + med + '\n')
        fileObj.write('The patient is advised to take ' + med + '\n')
        fileObj.write('The patient needs to take ' + med + '\n')
        fileObj.write(med + ' has been prescribed for the patient')
```
